```python
def handel_age(value):
    if(value=='NEW' or value=='_'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
ab['age']=ab['age'].apply(handel_age)
ab['age'].head()
```

```python
plt.subplots(figsize=(10,8))
sns.countplot(x = 'age', data = ab, palette='rainbow')
plt.title('AGE in List')
```

```python
fig = plt.figure(figsize=(20,10))
ax= fig.gca()

sns.boxplot(x = ab['age'], y = ab['G1'])
plt.show()
```

```python
sns.violinplot(x = ab['Mjob'], y = ab['age'])
plt.show()
```

```python
plt.figure(figsize=(15,5))
a=ab['absences'].value_counts()[:15]
sns.barplot(x=a,y=a.index,palette='rainbow')
plt.title('Number of absences')
plt.xlabel("Number of absences",size=15)
plt.ylabel("Number of absences",size=15)
```

```python
sns.countplot(ab['G1'])
sns.countplot(ab['absences']).set_xticklabels(sns.countplot(ab['G1']).get_xticklabels(),rotation=90)
fig = plt.gcf()
fig.set_size_inches(10,6)
plt.title('number of absences')
```
